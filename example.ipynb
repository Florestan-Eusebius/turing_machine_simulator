```python
import turing_machine_simulator

external_alphabet = ['0','1']
alphabet = external_alphabet + ['', '*', '0_', '1_']
states = {
    'q0' : 'initial state',
    'qf' : 'final state',
    'r0' : 'right moving, transferring 0',
    'r1' : 'right moving, transferring 1',
    'l0' : 'left moving, transferring 0',
    'l1' : 'left moving, transferring 1',
    'l0_' : 'left moving, transferring 0_',
    'l1_' : 'left moving, transferring 1_',
}
transition_function = {
    ('q0', '0') : ('r0', '*', 1),
    ('q0', '1') : ('r1', '*', 1),
    ('q0', ''): ('q0', '', -1),
    ('r0', '0') : ('r0', '0', 1),
    ('r0', '1') : ('r0', '1', 1),
    ('r1', '0') : ('r1', '0', 1),
    ('r1', '1') : ('r1', '1', 1),
    ('r0', '0_') : ('l0_', '0', -1),
    ('r0', '1_') : ('l0_', '1', -1),
    ('r0', '') : ('l0_', '', -1),
    ('r1', '0_') : ('l1_', '0', -1),
    ('r1', '1_') : ('l1_', '1', -1),
    ('r1', '') : ('l1_', '', -1),
    ('l0_', '0') : ('l0', '0_', -1),
    ('l0_', '1') : ('l1', '0_', -1),
    ('l1_', '0') : ('l0', '1_', -1),
    ('l1_', '1') : ('l1', '1_', -1),
    ('l0', '0') : ('l0', '0', -1),
    ('l0', '1') : ('l0', '1', -1),
    ('l1', '0') : ('l1', '0', -1),
    ('l1', '1') : ('l1', '1', -1),
    ('l0', '*') : ('q0', '0', 1),
    ('l1', '*') : ('q0', '1', 1),
    ('q0', '0_') : ('qf', '0', -1),
    ('q0', '1_') : ('qf', '1', -1),
    ('l0_', '*') : ('qf', '0', -1),
    ('l1_', '*') : ('qf', '1', -1),
}
TM = turing_machine_simulator.TuringMachine(alphabet, external_alphabet, states, transition_function)
TM.run_machine(['0','1','1','0','1','0','1','1','1','0','0'])
```

```python
external_alphabet = ['0','1']
alphabet = external_alphabet + ['', '+', '0_', '1_']
states = {
    'q0' : 'initial state',
    'qf' : 'final state',
}
transition_function = {
    # starting
    ('q0', '0') : ('sr+', '0_', 1),
    ('q0', '1') : ('sr+', '1_', 1),
    # moving right and fint +
    ('sr+', '0') : ('sr+', '0', 1),
    ('sr+', '1') : ('sr+', '1', 1),
    ('sr+', '+') : ('sl', '+', -1),
    # moving to the left to find the first non-labelled bit
    ('sl', '') : ('sl', '', -1),
    ('sl', '0') : ('a0', '0', 1),
    ('sl', '1') : ('a1', '1', 1),
    # moving right and find the position to add 
    ('a0', '0') : ('a0', '0', 1),
    ('a0', '1') : ('a0', '1', 1),
    ('a0', '0_') : ('a0_', '0', -1),
    ('a0', '1_') : ('a0_', '1', -1),
    ('a0', '') : ('a0_', '', -1),
    ('a1', '0') : ('a1', '0', 1),
    ('a1', '1') : ('a1', '1', 1),
    ('a1', '0_') : ('a1_', '0', -1),
    ('a1', '1_') : ('a1_', '1', -1),
    ('a1', '') : ('a1_', '', -1),
    # adding
    ('a0_', '0') : ('sl+', '0_', -1),
    ('a0_', '1') : ('sl+', '1_', -1),
}
TM = turing_machine_simulator.TuringMachine(alphabet, external_alphabet, states, transition_function)
TM.run_machine(['0','1','1','0','1','0','1','1','1','0','0'])
```
